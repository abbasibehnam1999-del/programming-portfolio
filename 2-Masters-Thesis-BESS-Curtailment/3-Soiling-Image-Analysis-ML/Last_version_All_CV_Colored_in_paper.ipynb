```python
# List of file paths and corresponding country names
file_paths = [
    (r"C:\Users\hp\Desktop\Results_1\Algerian\edited_main.xlsx", 'Algerian'),
    (r"C:\Users\hp\Desktop\Results_1\Australia\edited_main.xlsx", 'Australia'),
    (r"C:\Users\hp\Desktop\Results_1\Chile\edited_main.xlsx", 'Chile'),
    (r"C:\Users\hp\Desktop\Results_1\Jordanien\edited_main.xlsx", 'Jordanien'),
    (r"C:\Users\hp\Desktop\Results_1\Kap_Verde\edited_main.xlsx", 'Kap_Verde'),
    (r"C:\Users\hp\Desktop\Results_1\Moroko\edited_main.xlsx", 'Moroko'),
    (r"C:\Users\hp\Desktop\Results_1\Qatar\edited_main.xlsx", 'Qatar'),
    (r"C:\Users\hp\Desktop\Results_1\Spanien\edited_main.xlsx", 'Spanien'),
    (r"C:\Users\hp\Desktop\Results_1\USA\edited_main.xlsx", 'USA')
]

# Define colors for each country
country_colors = {
    'Algerian': 'blue',
    'Australia': 'green',
    'Chile': 'red',
    'Jordanien': 'purple',
    'Kap_Verde': 'orange',
    'Moroko': 'cyan',
    'Qatar': 'magenta',
    'Spanien': 'yellow',
    'USA': 'black'
}

# Define variable names to be displayed on x-axis
variables = [
    'Diam', 'Mean', 'Mode', 'Circ.', 'Feret', 'FeretX', 'FeretY',
    'FeretAngle', 'Minferet', 'AR', 'Round', 'Solidity', 'f', 'N'
]

# Initialize lists to store the data, labels, and colors
boxplot_data = []
labels = []
colors = []
positions = []

# Gap size between groups
gap_size = 2

def read_and_process_data(file_path, column_index, country_name):
    """Read and process data from the specified Excel file and column index."""
    try:
        df = pd.read_excel(file_path)
        if column_index == 14:
            column_data = df.iloc[:6, column_index].dropna()  # Column O data only up to row O6
        else:
            column_data = df.iloc[:, column_index].dropna()
        column_data = pd.to_numeric(column_data, errors='coerce').dropna()  # Ensure numeric data
        return column_data
    except Exception as e:
        print(f"Error processing file {file_path} for {country_name}: {e}")
        return pd.Series([])

# Loop through each column index (1 to 14 for B to O)
position = 1
for i in range(1, 15):
    # Collect data for the current column across all files
    for file_path, country_name in file_paths:
        column_data = read_and_process_data(file_path, i, country_name)
        if not column_data.empty:
            boxplot_data.append(column_data)
            labels.append(variables[i-1] if country_name == 'Algerian' else '')  # Display variable name only once per group
            colors.append(country_colors[country_name])
            positions.append(position)
            position += 1
    position += gap_size  # Add gap after each group

# Create a single combined box plot
fig, ax = plt.subplots(figsize=(24, 10))

# Plot each box plot individually with the specified color
boxplots = ax.boxplot(boxplot_data, positions=positions, patch_artist=True)

# Set colors for each box plot
for patch, color in zip(boxplots['boxes'], colors):
    patch.set_facecolor(color)

# Increase font size for X and Y axis labels
ax.set_xticks(positions)
ax.set_xticklabels(labels, rotation=90, ha='right', fontsize=12)

ax.set_title('', fontsize=16)
ax.set_xlabel('', fontsize=14)
ax.set_ylabel('CV(%)', fontsize=16)
ax.tick_params(axis='x', labelsize=13)
ax.tick_params(axis='y', labelsize=20)

# Add legend for countries
handles = [plt.Line2D([0], [0], color=color, lw=4) for color in country_colors.values()]
plt.legend(handles, country_colors.keys(), title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')

ax.grid(axis='y')
ax.set_ylim(0, 370)  # Set y-axis limit
plt.tight_layout()

# Save the plot
output_path = r"C:\Users\hp\Desktop\NNN\NNN.png"
plt.savefig(output_path)

# Show the plot
plt.show()
```
